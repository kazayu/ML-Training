Linear Regression

```python
import pandas as pd
import numpy as np
```

```python
salary_data = pd.read_csv('Salary_Data.csv')
salary_data
```

```python
salary_data.shape
```

```python
x= salary_data.iloc[:,:-1]
x
```

```python
y = salary_data.iloc[:,-1:]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,x_test,Y_train,y_test = train_test_split(x,y,test_size=0.20)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

Algorithm Selection

```python
from sklearn import linear_model
```

```python
lin_mod = linear_model.LinearRegression()#model selection
lin_mod.fit(X_train,Y_train)#training
```

```python
y_pred = lin_mod.predict(x_test)
y_pred
```

```python
y_test
```

Logistic Regression


```python
diabetes = pd.read_csv('diabetes.csv')
```

```python
diabetes.head()
```

Preparation of Data

```python
diabetes.shape
```

```python
input_x = diabetes.iloc[:,:-1]
input_x
```

```python
input_x.shape
```

```python
input_y = diabetes.iloc[:,-1:]
input_y
```

```python
X_train1,X_test1,y_train1,y_test1 = train_test_split(input_x,input_y,test_size = 0.15)
```

```python
print(X_train1.shape)
print(X_test1.shape)
print(y_train1.shape)
print(y_test1.shape)
```

```python
log_model = linear_model.LogisticRegression()#model_selection
log_model.fit(X_train1,y_train1)#training step
```

```python
y_pred = log_model.predict(X_test1)
y_pred
```

```python
y_test1
```

```python
from sklearn import metrics
```

```python
Accuracy = metrics.accuracy_score(y_pred,y_test1)
Accuracy
```
